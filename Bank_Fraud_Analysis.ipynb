```python
import pandas as pd
import numpy as np
```

```python
file = "Banking_Risk_Analytics_Dataset.xlsx"

customers = pd.read_excel(file, sheet_name="Customers")
accounts = pd.read_excel(file, sheet_name="Accounts")
transactions = pd.read_excel(file, sheet_name="Transactions")
fraud = pd.read_excel(file, sheet_name="FraudCases")
loans = pd.read_excel(file, sheet_name="Loans")
cards = pd.read_excel(file, sheet_name="Cards")
branches = pd.read_excel(file, sheet_name="Branches")
```

```python
def audit(df, name):
    print(f"\n{name}")
    print("-" * 40)
    print("Rows:", df.shape[0])
    print("Columns:", df.shape[1])
    print("Null Values:")
    print(df.isnull().sum())
    print("\nDuplicates:", df.duplicated().sum())
```

```python
audit(customers,"Customers")
audit(accounts,"Accounts")
audit(transactions,"Transactions")
audit(fraud,"FraudCases")
audit(loans,"Loans")
audit(cards,"Cards")
audit(branches,"Branches")
```
